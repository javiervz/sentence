## algunas cosas básicas de los datos

```python
def tokenize(s):
    return s.split(' ')
```

```python
import re

def remove_accents(raw_text):
    raw_text = re.sub(u"[àáâãäå]", 'a', raw_text)
    raw_text = re.sub(u"[èéêë]", 'e', raw_text)
    raw_text = re.sub(u"[ìíîï]", 'i', raw_text)
    raw_text = re.sub(u"[òóôõö]", 'o', raw_text)
    raw_text = re.sub(u"[ùúûü]", 'u', raw_text)
    return raw_text 
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

def clean(sesion):
    text=Datos_sesion[sesion]
    table = str.maketrans({key: None for key in '``!"#$%&\¿()*+,-./:;<=>?@[\\]_{|}'})
    text=[list(filter(None, [w.lower().translate(table) for w in tokenize(sentence)])) for sentence in text if len([w.lower() for w in [w.translate(table) for w in tokenize(sentence)]])>0]
    text=[[w for w in s if w!="''"] for s in text] 
    text=[[w for w in s if w!='̃'] for s in text] 
    text=[[remove_accents(w) for w in s if not w.isdigit()] for s in text]
    text=[[w for w in s if w not in stopwords.words('spanish')] for s in text] ## revisar esto!
    return text
```

```python
clean_text={}
for sesion in Datos_sesion.keys():
    C=clean(sesion)
    clean_text[sesion]=C
```

## graphs!

```python
import networkx as nx
```

```python
## adjacency graph
## n_sent: graph with n_sent sentences

def GoW(sesion,radius=4):
    G=nx.Graph()
    for sentence in clean_text[sesion]:
        if len(sentence)>1:
            pairs=[]
            for r in list(range(1,radius+1)):
                pairs+=list(zip(sentence,sentence[r:]))#+list(zip(sentence,sentence[2:]))+list(zip(sentence,sentence[3:]))
            for pair in pairs:
                if G.has_edge(pair[0],pair[1])==False:
                    G.add_edge(pair[0],pair[1],weight=1)
                else:
                    x=G[pair[0]][pair[1]]['weight']
                    G[pair[0]][pair[1]]['weight']=x+1
                    
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)      
    return G.subgraph(Gcc[0]).copy()
```

```python
graphs={}

for sesion in clean_text.keys():
    G=GoW(sesion)
    graphs[sesion]=G
```

```python
len(graphs)
```

### descomposición en cores

```python
core = {}
core_numbers = {}
for sesion in graphs.keys():
    G=graphs[sesion].copy()
    G.remove_edges_from(nx.selfloop_edges(G))
    core[sesion]=list(nx.k_core(G).nodes())
    core_numbers[sesion]=dict(nx.core_number(G))
```

```python
def sum_cores(core_numbers,node,G):
    N = list(dict(G[node]).keys())
    C = 0
    for nodenode in N:
        C+=core_numbers[nodenode]
    return C
```

```python
core_ordered = {}

for sesion in core.keys():
    G=graphs[sesion]
    core_ordered[sesion]=sorted({w:sum_cores(core_numbers[sesion],w,G) for w in core[sesion]}.items(), key=lambda x: x[1], reverse=True)
```
